```python
from skimage.metrics import structural_similarity
import cv2
import pandas as pd

#Works well with images of different dimensions
def orb_sim(img1, img2):
  # using ORB
  orb = cv2.ORB_create()

  # detect keypoints and descriptors
  kp_a, desc_a = orb.detectAndCompute(img1, None)
  kp_b, desc_b = orb.detectAndCompute(img2, None)

  # define the bruteforce matcher object
  bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    
  #perform matches. 
  matches = bf.match(desc_a, desc_b)
  #Look for similar regions with distance < 30. Goes from 0 to 100 so pick a number between.
  similar_regions = [i for i in matches if i.distance < 30]  
  if len(matches) == 0:
    return 0
  return len(similar_regions) / len(matches)


#Needs images to be same dimensions
def structural_sim(img1, img2):

  sim, diff = structural_similarity(img1, img2, full=True, multichannel=True)
  return sim


csv1 = pd.read_csv('products - Similarity.csv')
result = []
for j in range (0,csv1.shape[0]):
    for k in range (0,csv1.shape[0]):
        a = csv1.iloc[j][0]
        b = csv1.iloc[k][1]
        print(a)
        print(b)
        img1 = cv2.imread(f"{a}",1)
        img2 = cv2.imread(f"{b}",1)

        orb_similarity = orb_sim(img1, img2)  #1.0 means identical. Lower = not similar

        print("Similarity using ORB is: ", orb_similarity)

        #Resize for SSIM
        from skimage.transform import resize
        img5 = resize(img2, (img1.shape[0], img1.shape[1]), anti_aliasing=True, preserve_range=True)

        ssim = structural_sim(img1, img5) #1.0 means identical. Lower = not similar
        print("Similarity using SSIM is: ", ssim)

        if (((ssim + orb_similarity)/2)>=0.9 or orb_similarity>=0.9):
            print("Images are Similar")
            result.append(b)
    csv1.loc[j, 'Result'] = f"{result}"
    csv1.to_csv('products - Similarity.csv', index = False)
    result.clear()

    
```
